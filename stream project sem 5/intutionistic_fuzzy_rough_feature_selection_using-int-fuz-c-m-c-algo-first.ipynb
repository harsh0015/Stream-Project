```python
#data fuzzification using c means clustering algorithm:-
c=5#number of clusters
epsilon=0.000001


def calculate_cluster_center(mat,values_mem,values_non_mem,values_hes):
    center1=0
    div1=0
    center2=0
    div2=0
    center3=0
    div3=0
    v=[]
    for i in range(len(values_mem)):
        center1=center1+ values_mem[i]*mat[i]*mat[i]
        div1= div1 + mat[i]*mat[i]
    center1=center1/div1
    
    for i in range(len(values_mem)):
        center2=center2+ values_non_mem[i]*mat[i]*mat[i]
        div2= div2 + mat[i]*mat[i]
    center2=center2/div2
    
    for i in range(len(values_mem)):
        center3=center3+ values_hes[i]*mat[i]*mat[i]
        div3= div3 + mat[i]*mat[i]
    center3=center3/div3
    v.append(center1)
    v.append(center2)
    v.append(center3)
    return v
    
    

def fuzzify(values_mem,values_non_mem):
    print("hello")
    values_hes=[]
    for i in range(len(values_mem)):
        values_hes.append(1-values_mem[i]-values_non_mem[i])
    
    l=len(values_mem)
    
    #matrix contains the randomised numbers initially between 0 and 1 and the columns add up to 1
    
    membership_matrix=[]
   
    for j in range(c): 
        arr=[]
        for i in range(l):
            arr.append(random.random())
        membership_matrix.append(arr)
    
    for i in range(l):
        sum=0
        for j in range(c):
            sum=sum+membership_matrix[j][i]
        for j in range(c):
            membership_matrix[j][i]=membership_matrix[j][i]/sum
            
    for i in range(l):
        s=0
        for j in range(c):
            s=s+membership_matrix[j][i]
       
            
    t=100
    while t>0:
        
        t=t-1
        
        cluster_centers=[]
        for mat in membership_matrix:
            cluster_centers.append(calculate_cluster_center(mat,values_mem,values_non_mem,values_hes))
        new_membership_matrix=[]
        distance_matrix=[]
        
        for j in range(c):
            vec=[]
            for i in range(l):
                vec.append(math.sqrt(abs(values_mem[i]-cluster_centers[j][0])**2+abs(values_non_mem[i]-cluster_centers[j][1])**2+abs(values_hes[i]-cluster_centers[j][2])**2))
            distance_matrix.append(vec)
        
        for i in range(c):
            vec=[]
            for j in range(l):
                x=0;
                for k in range(c):
                    distance_matrix[i][j]=distance_matrix[i][j]+0.000001
                    distance_matrix[k][j]=distance_matrix[k][j]+0.000001
                    x=x+ (distance_matrix[i][j]/distance_matrix[k][j])*(distance_matrix[i][j]/distance_matrix[k][j])
                
                vec.append(1/x)
                
            new_membership_matrix.append(vec)
        
        flag=True
        for i in range(c):
            for j in range(l):
                if abs(membership_matrix[i][j]-new_membership_matrix[i][j])>epsilon:
                    flag=False
        if flag==False:
            membership_matrix=new_membership_matrix
            continue
        else :
            break
        
    fuzzified_values=[];
    for i in range(l):
        v=0
        for j in range(c):
            v=max(v,membership_matrix[j][i])
        fuzzified_values.append(v)
                
    
    
    return fuzzified_values
    
```
