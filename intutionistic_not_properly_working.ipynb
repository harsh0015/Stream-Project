```python
#data fuzzification using c means clustering algorithm:-
c=10 #number of clusters
epsilon=0.000001


def calculate_cluster_center(mat,values):
    center=0
    div=0
    for i in range(len(values)):
        center=center+ values[i]*mat[i]*mat[i]
        div= div + mat[i]*mat[i]
    center=center/div
    
    return center
    
    

def fuzzify(values):
    
    l=len(values)
    #matrix contains the randomised numbers initially between 0 and 1 and the columns add up to 1
    
    membership_matrix=[]
   
    for j in range(c): 
        arr=[]
        for i in range(l):
            arr.append(random.random())
        membership_matrix.append(arr)
    
    for j in range(c):
        sum=0
        for i in range(l):
            sum=sum+membership_matrix[j][i]
        for i in range(l):
            membership_matrix[j][i]=membership_matrix[j][i]/sum
    print(membership_matrix)
    t=100
    while t>0:
        
        t=t-1
        
        cluster_centers=[]
        for mat in membership_matrix:
            cluster_centers.append(calculate_cluster_center(mat,values))
        new_membership_matrix=[]
        distance_matrix=[]
        
        for j in range(c):
            vec=[]
            for i in range(l):
                vec.append(abs(values[i]-cluster_centers[j]))
            distance_matrix.append(vec)
        
        for i in range(c):
            vec=[]
            for j in range(l):
                x=0;
                for k in range(c):
                    x=x+ (distance_matrix[i][j]/distance_matrix[k][j])*(distance_matrix[i][j]/distance_matrix[k][j])
                
                vec.append(1/x)
                
            new_membership_matrix.append(vec)
        
        flag=True
        for i in range(c):
            for j in range(l):
                if abs(membership_matrix[i][j]-new_membership_matrix[i][j])>epsilon:
                    flag=False
        if flag==False:
            membership_matrix=new_membership_matrix
            continue
        else :
            break
        
    fuzzified_values=[];
    for i in range(l):
        v=0
        for j in range(c):
            v=max(v,membership_matrix[j][i])
        fuzzified_values.append(v)
                
    
    
    return fuzzified_values
    
```

```python
def diff(li1, li2):
    li_dif = [i for i in li1 + li2 if i not in li1 or i not in li2]
    return li_dif
```

```python
#reduct calculation for fuzzified data:-
lenn=len(df)
#function to calculate degree of dependence gamma:-

def calculate_rel(a,b,col):
    sd=np.std(col)
    if sd==0 :return 0
    return max(0,1-abs(a-b)/sd)
    
def calculate_gamma_intutionistic(R):
    
    
    vals_mem=[]
    vals_non_mem=[]
    for i in range(lenn):
        mi=10
        for attribute in R:
            mi=min(mi,df[attribute][i])
        vals_mem.append(mi)
        
    for i in range(lenn):
        mi=10
        for attribute in R:
            mi=min(mi,df_non[attribute][i])
        vals_non_mem.append(mi)
    
   
    pre=10
    lower_approx_mem=[]
    lower_approx_non_mem=[]
    
    
    
    for i in range(lenn):
        x=100
        for j in range(lenn):
            x=min(x,max(calculate_rel(vals_non_mem[i],vals_non_mem[j],vals_non_mem),vals_mem[j]))
        
        lower_approx_mem.append(x)
        
    for i in range(lenn):
        x=0
        for j in range(lenn):
            x=max(x,min(calculate_rel(vals_mem[i],vals_mem[j],vals_mem),vals_non_mem[j]))
        
        lower_approx_non_mem.append(x)
    
            
            
   
        
    sum=0;
    for i in range(lenn):
        sum+=(1+lower_approx_mem[i]-lower_approx_non_mem[i])/2
            
    return sum/lenn



    

def calculate_reduct():    
    R=[]
    gamma_best=0
    gamma_prev=-1;
    
    columns=['a','b','c','d','e','f','g','h','i','j','k','l','m']
    while gamma_best!=gamma_prev:
        
        T=R
        gamma_prev=gamma_best
        left=diff(columns,T)
        print(left)
        for attribute in left:
            new_gamma=calculate_gamma_intutionistic(R+[attribute])   
            print("gamma for:",R+[attribute],new_gamma)
            if(new_gamma>gamma_best):
                gamma_best=new_gamma
                T=R+[attribute]
        R=T
    
    return R



Reduct=calculate_reduct()
```
